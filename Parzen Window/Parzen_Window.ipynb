# Parzen Window Estimator

The main advantage of non-parametric density estimatio is that it allows us to estimate a Probability Density Function without an assumption of the distribution.

This notebook illustrates how desity estimation can be performed using the Parzen (Kernel) Window Estimator, which probably is the simplest of all the non-parametric estimators. <br>

The methods here called can be found in the *parzen_window.py* file

<img src="images/estimation.jpg" alt="Drawing" style="width: 350px;"/>


```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.ndimage as image
```

```python
from parzen_window import densityEstimation
```

## Loading and blurring image

The density that we want to estimate is the pixel amplitude of the Racoon image included in the scipy library.

```python
estimator = densityEstimation(sigma=5)
estimator.displayImages()
```

## Sampling from the density

In order to sample in an equitative and efficient method, we compute the Cumulative Density Function (CDF) and the map random numbers between 0 and 1 to the density value.

<img src="images/sampling.png" alt="Drawing" style="width: 400px;"/>

```python
numberOfSamples = 100000
samples, _ = estimator.drawSamples(numberOfSamples=numberOfSamples, from_where="training")
estimator.displayReconstruction( samples, numberOfSamples )
```
